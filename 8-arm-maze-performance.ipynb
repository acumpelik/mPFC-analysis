```python
import matplotlib.pyplot as plt

session1 = [50, 52, 58, 65, 69, 75] #red
session2 = [52, 60, 63, 70, 75, 80] #orange
session3 = [55, 68, 72, 75, 79, 82] #yellow
session4 = [63, 75, 80, 86, 88, 93] #green
session5 = [59, 80, 82, 89, 92, 93] #turquoise
session6 = [70, 87, 90, 96, 98, 97] #dark blue
session7 = [80, 97, 99, 100, 100, 100] #purple
trial = [1,2,3,4,5,6]

plt.plot(trial, session1, marker='o', color='red')
plt.plot(trial, session2, marker='o', color='darkorange')
plt.plot(trial, session3, marker='o', color='gold')
plt.plot(trial, session4, marker='o', color='limegreen')
plt.plot(trial, session5, marker='o', color='mediumturquoise')
plt.plot(trial, session6, marker='o', color='darkblue')
plt.plot(trial, session7, marker='o', color='purple')

plt.xlabel('Trial number', fontname='Arial', fontsize=14)
plt.ylabel('Expected performance (%)', fontname='Arial', fontsize=14)
plt.title('Expected learning curve', fontname='Arial', fontsize=20)

# plt.savefig('learning-curve-v.png', dpi=500)
```
