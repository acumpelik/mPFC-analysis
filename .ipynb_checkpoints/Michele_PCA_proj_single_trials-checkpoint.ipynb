```python
## visualize PCA projections of a population of neurons over trials ##
import numpy as np
import scipy.ndimage as nd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
## synthetic data - linear track running - details not important, just to play around ##

# generate 20 random firing fields on linear track 100 cm long
np.random.seed(12)
N=20
cents = np.random.rand(N)
x=np.linspace(0,1,100)
ws = np.random.randint(low=5,high=15,size=N)
frs = np.random.randint(low=5,high=15,size=N) + np.random.rand(N)
tun = np.array([frs[ic]*(1/4+np.exp(-ws[ic]*(x - c)**2)/2) for ic,c in enumerate(cents)])

# plot place cells
for i in range(N):
    plt.plot(tun[i])
plt.xlabel('position (cm)')
plt.ylabel('fr (sp/sec)')
plt.show()

# average speed for left trials
avg_spl = 20*np.exp(-30*(x - 0.5)**2) + 8
# average speed for right trials
avg_spr = 30 - avg_spl

# simulate running - let's say 20 times back and forths
locs = []
pos = 0
for tri in range(20):
    while pos<99:
        # jumps forward with a probability prop to avg_sp - let's say this is every 25ms (1/40)
        pos = pos + int(np.random.rand()<avg_spl[pos]/40)
        locs.append(pos)
    # random wait
    for i in range(np.random.randint(40, 400)):
        locs.append(pos)
    while pos>0:
        # jumps backward with a probability prop to avg_sp - let's say this is every 1ms
        pos = pos - int(np.random.rand()<avg_spr[pos]/40)
        locs.append(pos)
    # random wait
    for i in range(np.random.randint(40, 400)):
        locs.append(pos)

# poisson firing of each cell based on tuning
fir = np.random.poisson(tun[:,locs]/40)
```

```python
## check the size of the firing matrix
fir.shape
## this is 20 neurons, but each "unit" could also be the spikes detected from each indivdual tetrode.
## maybe check if these "mega units" have some spatial selectivity. Just plot their rate maps as if they were single units.
## HEADS UP: if you see that they are completely flat and/or super noisy, then probably there is no chance of extracting anything useful.
```

```python
## detect beginning / end of trials ##

### right-ward trials
# set start as the first time we are beyond 5cm
above_5 = (np.array(locs) > 5).astype(int)
r_starts = np.where(np.diff(above_5) == 1)[0]
# set end as the first time we are beyond 95cm
above_95 = (np.array(locs) > 95).astype(int)
r_ends = np.where(np.diff(above_95) == 1)[0]

### left-ward trials
# set start as the first time we are below 95cm
below_5 = (np.array(locs) < 95).astype(int)
l_starts = np.where(np.diff(below_5) == 1)[0]
# set end as the first time we are below 5cm
below_95 = (np.array(locs) < 5).astype(int)
l_ends = np.where(np.diff(below_95) == 1)[0]
```

```python
plt.plot(locs)
plt.xlabel('Time')
plt.ylabel('Location')
for l in l_starts:
    plt.scatter(l,100,s=1,c='k') # beginning of left-ward trials
for r in r_starts:
    plt.scatter(r,0,s=1,c='r') # beginning of right-ward trials
```

```python
fir.shape
```

```python
## plot raster for trial 1
raster=np.where(fir[:,:r_ends[0]])
plt.scatter(raster[1],raster[0],marker='|')
plt.xlabel('time window (25ms)')
plt.ylabel('neuron #')
plt.yticks(np.arange(5)*5-0.5,[1,5,10,15,20])
plt.show()
```

```python
## use PCA on entire firing dataset
pca = PCA(n_components=3)
pca.fit(fir.T)
print(pca.explained_variance_ratio_)
```

```python
## project firing onto first 2 principal components
proj = pca.transform(fir.T)
```

```python
proj.shape
```

```python
plt.plot(proj[:10000,0])
```

```python
## plot PCA for 5 left and 5 right trajectories

for i in range(5):
    plt.plot(proj[l_starts[i]:l_ends[i],0],proj[l_starts[i]:l_ends[i],1],color='blue',alpha=0.5)
    plt.plot(proj[r_starts[i]:r_ends[i],0],proj[r_starts[i]:r_ends[i],1],color='orange',alpha=0.5)
```

```python
### looks messy...let's try to smooth the firing a bit.
```

```python
fir_smooth = nd.gaussian_filter1d(fir.astype(float),50)
```

```python
pca_smooth = PCA(n_components=3)
pca_smooth.fit(fir_smooth.T)
print(pca_smooth.explained_variance_ratio_)
```

```python
## now the first PC explains >50% of the variance. Let's see if it looks better :)
```

```python
## project firing onto first 2 principal components
proj = pca_smooth.transform(fir_smooth.T)
```

```python
plt.plot(proj[:10000,0])
```

```python
## plot PCA for selected trajectories - let'd go fancy - 3D plot
ax = plt.axes(projection='3d')
for i in range(15):
    ax.plot3D(proj[l_starts[i]:l_ends[i],0],proj[l_starts[i]:l_ends[i],1],proj[l_starts[i]:l_ends[i],2],color='blue',alpha=0.5)
    ax.plot3D(proj[r_starts[i]:r_ends[i],0],proj[r_starts[i]:r_ends[i],1],proj[r_starts[i]:r_ends[i],2],color='orange',alpha=0.5)
ax.view_init(60, 85)
```

```python
## this looks much more reasonable. We see that, in PCA space, the neural activity "travels" from one location to another.
## moreover, we see a separation from left trials to right trials.
```
