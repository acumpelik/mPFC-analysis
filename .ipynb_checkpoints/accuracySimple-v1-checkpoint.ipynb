```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

accuracy = pd.read_csv('accuracySimpleData.csv')
```

```python
fig = plt.figure(figsize=(22,11))
ax = fig.add_subplot(111)
ax.set_title("Performance over time", fontsize=25)
ax.set_xlabel('Day', fontsize=18)
ax.set_ylabel('Accuracy', fontsize=18)
ax.tick_params(labelsize=18)
ax.xaxis.set_major_locator(mtick.MultipleLocator(1))
ax.yaxis.set_major_formatter(mtick.FormatStrFormatter('%.0f%%'))
ax.set_ylim(10,100)


linefmt = 1.5
linefmtHighlight = 4.5

# ax.plot(accuracy.Day, accuracy.JC240, c='red', label='JC240')
ax.plot(accuracy.Day[0:8], accuracy.JC240[0:8], c='red', linewidth=linefmt, label='JC240')
# ax.plot(accuracy.Day, accuracy.JC241, c='orange', linewidth=linefmt, label='JC241')
ax.plot(accuracy.Day[0:13], accuracy.JC241[0:13], c='orange', linewidth=linefmt, label='JC241')
ax.plot(accuracy.Day, accuracy.JC258, c='green', linewidth=linefmt, label='JC258')
# ax.plot(accuracy.Day[0:10], accuracy.JC258[0:10], c='green', linewidth=linefmt, label='JC258')
ax.plot(accuracy.Day, accuracy.JC267, c='blue', linewidth=linefmt, label='JC267')
ax.plot(accuracy.Day, accuracy.JC274, c='purple', linewidth=linefmtHighlight, label='JC274')
ax.yaxis.grid()

ax.legend(loc=2, fontsize=20)

plt.show()
```
