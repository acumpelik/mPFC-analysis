```python
import numpy as np
import matplotlib.pyplot as plt
import datetime
import matplotlib.dates as mdates
import matplotlib.ticker as mtick

weight_JC257 = [
    311,
    292,
    288,
    301,
    274,
    264,
    259
]

date_JC257 = [
    datetime.date(2021, 3, 22),
    datetime.date(2021, 3, 23),
    datetime.date(2021, 3, 24),
    datetime.date(2021, 3, 25),
    datetime.date(2021, 3, 26),
    datetime.date(2021, 3, 27),
    datetime.date(2021, 3, 28),
]

days = mdates.DayLocator() # find every day

amountfed_JC257 = [
    10,
    2.4,
    2.3,
    10,
    2.2,
    4.6,
    6,
]
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(date_JC257, weight_JC257, marker='o', markersize=10, color="black")
# ax.plot(weight_JC257)
ax.axhline(264, color='#cc0000', linewidth='3')
ax.axhline(249, color='#a6a6a6', linewidth='3', linestyle=':', alpha=0.5)
# ax.set_ylim(240, 320)
ax.set_ylim(233.25, 326.55)
# ax.set_xlabel("Date", fontsize=15)
ax.set_ylabel("Weight (g)", fontsize=15)
ax.tick_params(axis='x', labelsize=15, rotation=50)
ax.tick_params(axis='y', labelsize=15)
ax.xaxis.set_major_locator(days) # set ticks to plot each day

ax2 = ax.twinx()
ax2.set_ylabel("% of original weight", fontsize=15)
ax2.set_ylim(75, 105)
ax2.tick_params(axis='y', labelsize=15)
fmt = '%.0f%%'
yticks = mtick.FormatStrFormatter(fmt)
ax2.yaxis.set_major_formatter(yticks)

ax2.yaxis.grid()

plt.show()
# plt.savefig("Weight_JC257.jpg")

fig2, ax = plt.subplots(figsize=(12,4))
ax.bar(date_JC257, amountfed_JC257, alpha=0.3, color=('k','b','b','k','b','b','b'))
ax.set_ylabel("Amount fed (g)", fontsize=15)
ax.tick_params(axis='y', labelsize=15)
plt.show()
```
