```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import datetime

accuracy = pd.read_csv('accuracySimpleData.csv')
```

```python
fig = plt.figure(figsize=(22,11))
ax = fig.add_subplot(111)
ax.set_title("Performance over time", fontsize=25)
ax.set_xlabel('Day', fontsize=18)
ax.set_ylabel('Accuracy', fontsize=18)
ax.tick_params(labelsize=18)
ax.xaxis.set_major_locator(mtick.MultipleLocator(1))
ax.yaxis.set_major_formatter(mtick.FormatStrFormatter('%.0f%%'))
ax.yaxis.grid()
ax.axhline(y=80, c='black', ls='--') # color formerly #ae2012
ax.set_ylim(10,100)


linefmt = 1.5
linefmtHighlight = 3.5
cscheme = ["#2a9d8f", "#7a5195", "#ffa600", "#e76f51"]
markerfmt = 5
markerfmtHighlight = 6

# ax.plot(accuracy.Day, accuracy.JC240, c='red', label='JC240', marker='o', markersize=markerfmt)
ax.plot(accuracy.Day[0:8], accuracy.JC240[0:8], c='0.8', linewidth=linefmt, label='JC240', marker='o', markersize=markerfmt)
# ax.plot(accuracy.Day, accuracy.JC241, c='blue', linewidth=linefmt, label='JC241', marker='o', markersize=markerfmt)
ax.plot(accuracy.Day[0:13], accuracy.JC241[0:13], c='0.8', linewidth=linefmt, label='JC241', marker='o', markersize=markerfmt)
# ax.plot(accuracy.Day, accuracy.JC258, c=cscheme[0], linewidth=linefmt, label='JC258', marker='o', markersize=markerfmt)
ax.plot(accuracy.Day[0:10], accuracy.JC258[0:10], c='green', linewidth=linefmt, label='JC258', marker='o', markersize=markerfmt)
ax.plot(accuracy.Day, accuracy.JC267, c=cscheme[1], linewidth=linefmt, label='JC267', marker='o', markersize=markerfmt)
ax.plot(accuracy.Day[0:11], accuracy.JC274[0:11], c=cscheme[2], linewidth=linefmtHighlight, label='JC274', marker='o', markersize=markerfmtHighlight)
# ax.plot(accuracy.Day, accuracy.JC274, c=cscheme[2], linewidth=linefmtHighlight, label='JC274', marker='o', markersize=markerfmtHighlight)

ax.legend(loc=2, fontsize=20, framealpha=1)

plt.show()
timestamp = datetime.datetime.today().strftime("%Y-%m-%d-%H%M%S")
# plt.savefig("Performance_"+timestamp+".png")
```
