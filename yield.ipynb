```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
unit_yield = pd.read_csv('data/yield.csv', usecols=[0,2,3,4])
```

```python
unit_yield.boxplot(by=['Animal'])
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.subplot(1,2,1)
sns.violinplot(x = unit_yield['Animal'], y=unit_yield['PFC'])

plt.subplot(1,2,2, sharey=plt.gca())
boxplot = sns.violinplot(x = unit_yield['Animal'], y=unit_yield['HPC'])
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.subplot(1,2,1)
sns.boxplot(x = unit_yield['Animal'], y=unit_yield['PFC'])

plt.subplot(1,2,2, sharey=plt.gca())
boxplot = sns.boxplot(x = unit_yield['Animal'], y=unit_yield['HPC'])
plt.show()
```
