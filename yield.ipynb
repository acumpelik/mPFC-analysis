```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import datetime
```

```python
unit_yield = pd.read_csv('data/yield.csv', usecols=[0,2,3,4])
timestamp = datetime.datetime.today().strftime("%Y-%m-%d-%H%M%S")
```

```python
plt.figure(figsize=(10,6))
plt.suptitle('Unit yield for PFC and HPC')

# PFC
plt.subplot(1,2,1)
pfc_plot = sns.violinplot(x = unit_yield['Animal'], y=unit_yield['PFC'])
pfc_plot.set(ylabel="PFC unit yield", title="PFC")

# HPC
plt.subplot(1,2,2, sharey=plt.gca())
plt.title('HPC')
hpc_plot = boxplot = sns.violinplot(x = unit_yield['Animal'], y=unit_yield['HPC'])
hpc_plot.set(ylabel="HPC unit yield")

plt.savefig("yield_violinplot_"+timestamp+".png")
# plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.suptitle('Unit yield for PFC and HPC')

# PFC
plt.subplot(1,2,1)
pfc_plot = sns.boxplot(x = unit_yield['Animal'], y=unit_yield['PFC'])
pfc_plot.set(ylabel='units', title='PFC')

# HPC
plt.subplot(1,2,2, sharey=plt.gca())
hpc_plot = sns.boxplot(x = unit_yield['Animal'], y=unit_yield['HPC'])
hpc_plot.set(ylabel='units', title='HPC')

plt.savefig("yield_boxplot_"+timestamp+".png")
# plt.show()
```
